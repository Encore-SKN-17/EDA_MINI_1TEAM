```python
import matplotlib.pyplot as plt

df_music2['release_year'] = pd.to_datetime(df_music2['release_date'], errors='coerce').dt.year

df_movie = df_movie.rename(columns={'Year': 'release_year'})

movie_by_year = df_movie.groupby('release_year')['Duration_minute'].mean().reset_index()
movie_by_year.rename(columns={'release_year': 'year', 'Duration_minute': 'avg_movie_duration'}, inplace=True)

music_by_year = df_music2.groupby('release_year')['time_signature'].mean().reset_index()
music_by_year.rename(columns={'release_year': 'year', 'time_signature': 'avg_music_duration'}, inplace=True)

total = pd.merge(movie_by_year, music_by_year, on='year', how='inner')

# 정렬
total = total.sort_values('year')

total = total.sort_values('year')

fig, ax1 = plt.subplots(figsize=(12, 6))

ax1.plot(total['year'], total['avg_movie_duration'], color='blue', marker='o', label='영화 러닝타임')
ax1.set_ylabel('영화 러닝타임 (분)', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')

ax2 = ax1.twinx()
ax2.plot(total['year'][::-1], total['avg_music_duration'], color='red', marker='s', label='음악 길이 (반대방향)')
ax2.set_ylabel('음악 길이 (분)', color='red')
ax2.tick_params(axis='y', labelcolor='red')

plt.title('연도별 러닝타임 vs 연도별 음악 러닝타임')
ax1.set_xlabel('연도')
ax1.grid(True)

lines_1, labels_1 = ax1.get_legend_handles_labels()
lines_2, labels_2 = ax2.get_legend_handles_labels()
plt.legend(lines_1 + lines_2, labels_1 + labels_2, loc='upper left')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df_music2['release_year'] = pd.to_datetime(df_music2['release_date'], errors='coerce').dt.year

df_movie = df_movie.rename(columns={'Year': 'release_year'})

# IQR 기준 이상치 제거
Q1 = df_movie['Duration_minute'].quantile(0.25)
Q3 = df_movie['Duration_minute'].quantile(0.75)
IQR = Q3 - Q1
df_movie_filtered = df_movie[(df_movie['Duration_minute'] >= Q1 - 1.5 * IQR) & (df_movie['Duration_minute'] <= Q3 + 1.5 * IQR)]

# 연도별 평균 영화 러닝타임 계산
movie_by_year = df_movie_filtered.groupby('release_year')['Duration_minute'].mean().reset_index()
movie_by_year.rename(columns={'release_year': 'year', 'Duration_minute': 'avg_movie_duration'}, inplace=True)

# 연도별 평균 음악 길이 계산
music_by_year = df_music2.groupby('release_year')['time_signature'].mean().reset_index()
music_by_year.rename(columns={'release_year': 'year', 'time_signature': 'avg_music_duration'}, inplace=True)

total = pd.merge(movie_by_year, music_by_year, on='year', how='inner')
total = total.sort_values('year')

fig, ax1 = plt.subplots(figsize=(12, 6))

ax1.plot(total['year'], total['avg_movie_duration'], color='blue', marker='o', label='영화 러닝타임')
ax1.set_ylabel('영화 러닝타임 (분)', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')

ax2 = ax1.twinx()
ax2.plot(total['year'][::-1], total['avg_music_duration'], color='red', marker='s', label='음악 길이 (반대방향)')
ax2.set_ylabel('음악 길이 (분)', color='red')
ax2.tick_params(axis='y', labelcolor='red')

plt.title('연도별 영화 러닝타임 vs 음악 길이')
ax1.set_xlabel('연도')
ax1.grid(True)

lines_1, labels_1 = ax1.get_legend_handles_labels()
lines_2, labels_2 = ax2.get_legend_handles_labels()
plt.legend(lines_1 + lines_2, labels_1 + labels_2, loc='upper left')

plt.tight_layout()
plt.show()
```
